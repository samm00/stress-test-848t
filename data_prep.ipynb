## Load Data

#### Folk Tales

```python
df_folk_tales = pd.DataFrame(columns=['trg', 'eng', 'tale', 'chapter'])

with open('data/kachalona.txt') as f: 
    text = f.read()
    text = text.split('\n-----\n\n')
    for idx, chapter in enumerate(text):
        chapter = chapter.split('\n')
        text_trg = chapter[::3]
        text_eng = chapter[1::3]

        n = len(text_trg)

        new_chapter_df = pd.DataFrame(zip(text_trg, text_eng, ['kachalona']*n, [idx+1]*n), columns=['trg', 'eng', 'tale', 'chapter'])
        df_folk_tales = pd.concat([df_folk_tales, new_chapter_df])

df_folk_tales
```

```python
df_malkammot = pd.read_csv('data/malkammot.csv')

df_malkammot['tale'] = 'malkammot'
df_malkammot['chapter'] = [1 if x < 20 else 2 for x in list(df_malkammot.index)]

df_folk_tales = pd.concat([df_folk_tales, df_malkammot])
df_malkammot.head()
```

```python
print(f'Total sentences: {len(df_folk_tales)}')
df_folk_tales.to_csv('data/folk_tales.csv', index=False)
```
